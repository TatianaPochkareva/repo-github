```python
# задача 3,1 Построение графика окружности.
x = []
x2 = []
y = []
y2 = []

for i in range(1500):
    r = 1000
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
    
plt.figure(figsize = (6, 6))    
plt.plot(x, y, color='r')
plt.plot(x, y2, color='r')
plt.plot(x2, y2, color='r')
plt.plot(x2, y, color='r')
plt.axis('scaled')
plt.show()
```

```python
#  Задание 4 Решите систему уравнений: y = x2 – 1  exp(x) + x(1 – y) = 1;
# y = x2 – 1    y = (exp(x) + x – 1) / x

x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
#  Задание 4.2 Решите систему уравнений и неравенств:y = x2 – 1   exp(x) + x * (1 – y) > 1
# y = x2 – 1  y = -1/x +2


x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
